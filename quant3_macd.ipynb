Here, we are generating Buying signals based on MACD.
for this, we calculate 12ema and 26ema, macd line which is 12ema-26ema

Then we calculated 9 ema of macd, which is also called signal line

Buy signal:
When macd line is above 9 ema

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df = yf.download('^NSEBANK', '2006-01-01')
```

```python
df
```

```python
df['ema12'] = df['Close'].ewm(span=12).mean()
```

```python
df['ema26'] = df['Close'].ewm(span=26).mean()
```

```python
df
```

```python
df['macd'] =  df['ema12'] - df['ema26']
```

```python
df['9ema'] = df['macd'].ewm(span=9).mean()
```

```python
df2 = df.drop([ 'Open', 'High', 'Low', 'Volume', 'Adj Close'   ] , axis = 'columns'   )
```

```python
df2
```

```python
plt.figure (figsize=(20,8))
plt.plot(df['Close'])
plt.plot(df['macd'])
plt.plot(df['9ema'])

plt.show()
```

```python
buy = []
sell = []

for i in range(len(df2)):
  if (df['macd'].iloc[i] > df['9ema'].iloc[i]) and df['macd'].iloc[i-1]< df['9ema'].iloc[i-1]:
    buy.append(i)
```

```python
len(buy)
```

```python
plt.figure(figsize=(20,8))
plt.plot(df['Close'], label = 'Close', c='blue', alpha =0.7 )
plt.plot(df['macd'], label = 'macd',c ='red', alpha = 0.7)
plt.plot(df['9ema'], label = '9ema', c ='green', alpha = 0.7)

plt.scatter(df.iloc[buy].index, df.iloc[buy].Close , marker ='^', c ='green', s=100)
```

In the above figure, green symbol denotes buy signals
