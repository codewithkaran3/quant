In this code, we are trying to understand the movement of stock price.

That is, out of 100 trading days, for how many days the stock traded between -1 to 1 percentage, more than 2%, etc

This information will help intraday traders, to select a high beta stock

```python
import pandas as pd
import yfinance as yf

df = yf.download('SBIN.NS', start = '2008-01-01')
```

```python
df
```

```python
df['price_change'] = df['Close'].pct_change()
```

```python
df
```

Calculating percent column by multiplying column 'Price_change' with 100

```python
df['percent'] = df['price_change']*100
```

```python
df
```

```python
def analyze(r):
  less_one = []
  one_two = []
  two_three = []
  three_plus = []

  for i in range(len(r)):
    if (r.iloc[i]>=0 and r.iloc[i] <1) or (r.iloc[i]>-1 and r.iloc[i]<0):
      less_one.append(i)
    elif (r.iloc[i]>=1 and r.iloc[i] <2) or (r.iloc[i]>-2 and r.iloc[i]<-1):
      one_two.append(i)
    elif (r.iloc[i]>=2 and r.iloc[i] <3) or (r.iloc[i]>-3 and r.iloc[i]<-2):
      two_three.append(i)

    elif (r.iloc[i]>=3) or (r.iloc[i]<-3):
      three_plus.append(i)


  def count():
    l = len(df)
    o = len(less_one)
    t = len(one_two)
    th = len(two_three)
    thp = len(three_plus)

    print(o, 'is less than one percentage in', l ,"days", 'so the percentage is ', o/l)
    print(t, 'is between one and two percentage in', l ,"days", 'so the percentage is ', t/l)
    print(th, 'is between two and three in', l, "days", 'so the percentage is ', th/l)
    print(thp, 'is more than three percentage in', l, "days", 'so the percentage is ', thp/l)
 
  count()
```

Results:

```python
r = df['percent']
analyze(r)
```
